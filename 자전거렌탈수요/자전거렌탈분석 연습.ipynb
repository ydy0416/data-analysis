```python
import pandas as pd
```

```python
train_df=pd.read_csv('train.csv')
test_df=pd.read_csv('test.csv')
```

```python
#열 삭제 
train_df.drop(['casual', 'registered'], axis=1, inplace=True)
```

```python
train_df['year']=train_df['datetime'].str[:4]
train_df['month']=train_df['datetime'].str[5:7]
train_df['day']=train_df['datetime'].str[8:11]
train_df['hour']=train_df['datetime'].str[11:13]

test_df['year']=test_df['datetime'].str[:4]
test_df['month']=test_df['datetime'].str[5:7]
test_df['day']=test_df['datetime'].str[8:11]
test_df['hour']=test_df['datetime'].str[11:13]
```

```python
train_df.drop(['datetime'], axis=1, inplace=True)
```

```python
train_df
```

```python
train_df.info()
```

```python
X_train=train_df.drop(['count', 'year'], axis=1)
y_train=train_df['count']
X_test=test_df.drop(['datetime', 'year'], axis=1, inplace=False)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor(n_estimators=100)
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred=rf.predict(X_test)
```

```python
y_pred
```

```python
submission=pd.DataFrame({"datetime":test_df['datetime'],"count":y_pred})
```

```python
submission
```
