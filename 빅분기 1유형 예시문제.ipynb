```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/dataq/main/mtcars.csv',index_col=0)
```

```python
X=df['qsec']
X_MinMax=(X-X.min(axis=0))/(X.max(axis=0)-X.min(axis=0))
overNum=len(X_MinMax[X_MinMax>0.5])
```
